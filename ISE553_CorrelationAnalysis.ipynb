## Correlation Matrix

```python
# Import the libraries
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")
```

```python
# read data
path = r'C:\Users\thasnai\Documents\Python\Book1.xlsx'
df = pd.read_excel(path)
df.head()
```

```python
# Create dataframe for 15 lags
Col_Names = ["Lag_1","Lag_2","Lag_3","Lag_4","Lag_5","Lag_6","Lag_7","Lag_8","Lag_9","Lag_10","Lag_11","Lag_12","Lag_13","Lag_14","Lag_15"]
index = np.arange(106)
Corr_df = pd.DataFrame(index = index, columns=Col_Names)
Corr_df["Lag_1"] = df[0:106].to_numpy()
Corr_df["Lag_2"] = df[1:107].to_numpy()
Corr_df["Lag_3"] = df[2:108].to_numpy()
Corr_df["Lag_4"] = df[3:109].to_numpy()
Corr_df["Lag_5"] = df[4:110].to_numpy()
Corr_df["Lag_6"] = df[5:111].to_numpy()
Corr_df["Lag_7"] = df[6:112].to_numpy()
Corr_df["Lag_8"] = df[7:113].to_numpy()
Corr_df["Lag_9"] = df[8:114].to_numpy()
Corr_df["Lag_10"] = df[9:115].to_numpy()
Corr_df["Lag_11"] = df[10:116].to_numpy()
Corr_df["Lag_12"] = df[11:117].to_numpy()
Corr_df["Lag_13"] = df[12:118].to_numpy()
Corr_df["Lag_14"] = df[13:119].to_numpy()
Corr_df["Lag_15"] = df[14:120].to_numpy()
Corr_df.head()
```

```python
# Calculate and create correlation

# Compute the correlation matrix
corr = Corr_df.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
corr_diagram = sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

ax.set_title('Diagonal Correlational Matrix', fontsize = 20)

fig = corr_diagram.get_figure()
fig.savefig('Output.png') 
```
